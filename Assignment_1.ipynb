## Analyze Audio using Librosa

```python
!pip install librosa
```

```python
import librosa
audio_path = "/content/Assignment.mp3"
x, sr = librosa.load(audio_path)
```

```python
import IPython.display as ipd
ipd.Audio(audio_path)
```

```python
#WAVEPLOT
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize = (12,8))
librosa.display.waveshow(x, sr = sr)
plt.show()
```

```python
#SPECTROGRAM
X = librosa.stft(x) # short term fourier transform
Xdb = librosa.amplitude_to_db(abs(X)) #amplitude at any given time
```

```python
plt.figure(figsize = (12,8))
#librosa.display.specshow(Xdb, sr=sr, x_axis = 'time',y_axis = 'hz')
librosa.display.specshow(Xdb, sr=sr, x_axis = 'time',y_axis = 'log')
plt.colorbar()
```

Output suggests that there's not much gap in between the speech and all follow a single pitch. Some words are extremely long to spell so we can see a contiuous long waves of same tones
